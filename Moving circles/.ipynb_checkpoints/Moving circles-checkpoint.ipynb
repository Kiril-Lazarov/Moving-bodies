```python
import numpy as np 
import pygame 
import time
import math
```

```python
def update_body_coords(body,FPS,frame_number):
    rotation_angle = (2*np.pi  / FPS)* frame_number
    body['pos'][0] = np.cos(rotation_angle)
    body['pos'][1] = np.sin(rotation_angle)
   
    
    return [body['pos'][0], body['pos'][1]]
    
```

```python
def bodies_animation(screen_width=1500, screen_height=800):
    pygame.init()
    
    center = [screen_width/2, screen_height/2]
    
    body1 = {'color': (0,0,250),
                 'pos': center,
                 'radius': 10}
    body2 = {'color': (0,0,250),
                 'pos': [center[0]+100, center[1]],
                 'radius': 10}
    bg_color = (255, 255, 255)
    win = pygame.display.set_mode((screen_width, screen_height))
    # win.blit(background_image, (0, 0))

    clock = pygame.time.Clock()
    
    # Frames per second
    FPS = 10  
    clock.tick(FPS)
    running = True
    frame = 0
    # pygame.draw.circle(win, body2['color'], body2['pos'], body2['radius']) 
    while running:
        frame += 1
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False
                
        win.fill(bg_color)
        body2['pos'][0], body2['pos'][1] = update_body_coords(body2, FPS,frame)
        print('Body x: ', body['pos'][0])
        print('Body y: ', body['pos'][1])
        pygame.draw.circle(win, body1['color'], body1['pos'], body1['radius'])
        
        
        pygame.draw.circle(win, body2['color'], body2['pos'], body2['radius']) 
        pygame.display.update()
        pygame.display.flip()
    # Exit
    pygame.quit()
    
```
